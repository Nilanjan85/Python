## Flight Price Prediction - Day 3 (EDA + Feature Engineering)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sn
%matplotlib inline
```

```python
train_df = pd.read_excel("Flight_predtion_Train.xlsx")
train_df.head()
```

```python
test_df = pd.read_excel("Flight_predtion_Test.xlsx")
test_df.head()
```

```python
df = train_df.append(test_df)
```

```python
## Feature Engineering
## Create 3 columns for day month & year  from date of journey column 

df["Day"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]
df.head()
```

```python
## convert to int
df["Day"] = df["Day"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)
```

```python
df.info()
```

```python
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Arrival_Time"].str[0:5]
```

```python
df["Arrival_Time"] = df['Arrival_Time'].str.split(" ").str[0]
```

```python
df["Arrival_Hr"] = df['Arrival_Time'].str.split(":").str[0]
df["Arrival_Min"] = df['Arrival_Time'].str.split(":").str[1]
df.head()
```

```python
df["Arrival_Hr"] = df["Arrival_Hr"].astype(int)
df["Arrival_Min"] = df["Arrival_Min"].astype(int)
```

```python
df.info()
```

```python
df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
df.head(1)
```

```python
## Departure time
df["Dept_Hr"] = df['Dep_Time'].str.split(":").str[0]
df["Dept_Min"] = df['Dep_Time'].str.split(":").str[1]
df["Dept_Hr"] = df["Dept_Hr"].astype(int)
df["Dept_Min"] = df["Dept_Min"].astype(int)
df.drop("Dep_Time",axis=1,inplace=True)
df.head(1)
```

```python
df.info()
```

```python
df[df["Total_Stops"].isnull()]
```

```python
## Total Stops 

df["Total_Stops"].unique()
```

```python
## Find out what to substiute in NaN value for Total stops. We have to look for the same destination and source route and 
## find the most no of stops and will that value to substitute. 
df[(df["Source"] == "Delhi") & (df["Destination"] == "Cochin")]["Total_Stops"].value_counts()
```

```python
# Conclusion: The most no of stop is 1. 
```

```python
df["Total_Stops"] = df["Total_Stops"].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
#df["Total_Stops"] = df["Total_Stops"].astype(int)
df.drop("Route",axis=1,inplace=True)
```

```python
df.head()
```

```python
## duration
#Extract the hrs from duration
df["Dur_Hr"] = df["Duration"].str.split(" ").str[0].str.split("h").str[0]

# To convert the it to int got error as duration was 5m then we remove those records. 
df.drop(df[df["Dur_Hr"] == '5m'].index,axis=0,inplace=True)
```

```python
df["Dur_Hr"].astype(int)
```

```python
#Extract the min from duration
df["Dur_min"] = df["Duration"].str.split(" ").str[1].str.split("m").str[0].fillna(0).astype(int)
```

```python
# Replace the blank min with 0 
df["Dur_min"] = df["Dur_min"].replace('',0)
```

```python
df.head(2)
```

```python
df.drop("Duration",axis=1,inplace=True)
```

```python
df.head(2)
```

```python
# Categorical data needs to be handled 

df["Airline"].unique()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df["Airline"] = le.fit_transform(df["Airline"])
df["Source"] = le.fit_transform(df["Source"])
df["Destination"] = le.fit_transform(df["Destination"])
```

```python
df["Additional_Info"] = le.fit_transform(df["Additional_Info"])
```
