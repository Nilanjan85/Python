```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
### Corealtion with Heatmap
```

```python
# A corelation heatmap uses coloured cell, typically in a monochormatic scale to show a 2d corelation matrix(table) 
#between two dim or event types. its very important in Feature Selection
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
### JoinPlot
### <p> A join plot allows to study the relationship between 2 numeric variables. The central chart display their corelation. It is usually a scatterplot, a hexbin plot, a 2d histogram or a 2d density plot.</p>   
```

```python
### Bivariate Analysis
```

```python
sns.jointplot(x='tip',y ='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y ='total_bill',data=df,kind='reg')
```

```python
### Pair plot
```

```python
###<div>A pairplot is also know as a scatterplot in which one variable in the same data row is matched with another variables value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables </div>
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

```python
### Dist plot
```

```python
### dist plot helps us to check the distribution of the columns feature 
```

```python
sns.distplot(df['tip']) # histogram 
```

```python
sns.distplot(df['tip'],kde=False,bins=18) # histogram only showing the count not the pdf
```

```python
### Ctegorial Plots
```

```python
### Seaborn also helps us in doing the analysis on Categorial data points.
```

```python
### <ol><li>boxplot</li> <li>violinplot</li> <li>countplot</li> <li>bar plot</li> </ol>   
```

```python
## count plot

sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
# bar plot

sns.barplot(y='total_bill',x='sex',data=df)
```

```python
### box plot
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```
