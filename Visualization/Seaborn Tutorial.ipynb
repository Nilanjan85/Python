## Distribution Plot

### Distplot - for 2 features for Bivariate Analysis 
### joinplot - for 2 feature for Bivariate Analysis 
### pairplot - for more that 2 features multivariate analysis

```python
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
```

### Correlation with Heatmap 

```python
#calculate the correlation of the dataset 

df.corr()
```

```python
# Construct the heat map with the correlational data 

sns.heatmap(df.corr()) # for a heatmap we have to provide a input to it a matrix 
```

```python
# Joint plot for the bivariate analysis 
sns.jointplot(x="tip",y="total_bill",data=df,kind="hex");
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="reg");
```
