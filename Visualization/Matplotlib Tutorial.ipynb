## Matplotlib library

```python
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

## Scatter Plot

```python
plt.scatter(x,y,c='b')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Simple Scatter plot")
plt.savefig("scatter_simple.png")
```

## Plot

```python
y=x*x
```

```python
#plt plot

#plt.plot(x,y,'r') # x,y,colour n pattern 
#plt.plot(x,y,'r--')
#plt.plot(x,y,'r*--')
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('plot diagram')
```

## Subplot

```python
# creating subplot 
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'g')
plt.subplot(2,2,4)
plt.scatter(x,y,c='r')
```

```python
# subplot plotting of sin and cos graph 

x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)


plt.subplot(1,2,1)
plt.plot(x,y_sin,'r')
plt.title('Sine plot')


plt.subplot(1,2,2)
plt.plot(x,y_cos,'g')
plt.title('Cos plot')
```

## Bar Plot

```python
# creating bar plots

x = [2,8,6]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')

plt.title('Bar Plot')
plt.xlabel(' X axis')
plt.ylabel(' Y axis')
```

## Histogram

```python
x = [22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]

plt.hist(x)
plt.title("Histogram")
```

## Box Plot using matplotlib 

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

#rectangular box plot

plt.boxplot(data,vert=True,patch_artist=True);
```

## Pie Chart 

```python
label = 'Python','C++','Ruby','Java'
size = [215,130,245,210]
colours = ['gold','yellowgreen','lightcoral','lightskyblue']

plt.pie(size,labels=label,colors=colours,autopct='%1.1f%%',shadow=True);
```

```python
# explode

label = 'Python','C++','Ruby','Java'
size = [215,130,245,210]
colours = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = [0.2,0,0,0]

plt.pie(size,labels=label,explode=explode,colors=colours,autopct='%1.1f%%',shadow=True);

```
