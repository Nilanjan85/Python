# Probability Distrubution- Generating the random variable of various distrubution. 

```python
import seaborn as sns 
import scipy.stats as stats
```

```python
# 1. Uniform Disrtibution
n=10000
start=0
width=10
uni_dist = stats.uniform.rvs(size=n,loc=start,scale=width)

ax=sns.distplot(uni_dist,bins=100,kde=True,color='skyblue',
                hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Uniform Distribution',ylabel='Frequency')

ax.plot()
```

```python
# 2. Normal Distrubution

norm_dist = stats.norm.rvs(size=10000,loc=0,scale=1)


ax=sns.distplot(norm_dist,bins=100,kde=True,color='skyblue',
                hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution',ylabel='Frequency')
```

```python
#3. Bernoulli Distribution
bernouli_dist = stats.bernoulli.rvs(size=10000,p=0.6)

ax=sns.distplot(bernouli_dist,bins=100,kde=False,color='skyblue',
                hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Bernoulli Distribution',ylabel='Frequency')
```

```python
#4. Binomial Distribution

binom_dist = stats.binom.rvs(size=10000,n=10,p=0.8)

ax=sns.distplot(binom_dist,bins=100,kde=False,color='skyblue',
                hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Binomial Distribution',ylabel='Frequency')
```

```python
# 5. Poission Distribution

poission_dist= stats.poisson.rvs(mu=3,size=10000)


ax=sns.distplot(poission_dist,bins=100,kde=True,color='skyblue',
                hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Poission Distribution',ylabel='Frequency')
```

```python
# 6. Exponential Distribution
expon_dist = stats.expon.rvs(size=1000,scale=1,loc=0)

ax=sns.distplot(expon_dist,bins=100,kde=True,color='skyblue',
                hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Exponential Distribution',ylabel='Frequency')
```
