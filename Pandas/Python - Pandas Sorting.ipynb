```python
people = { 'first':['ram','shyam','jadu'],
    'last' : ['a','a','c'],
    'email' : ['ram@email.com','shyam@email.com','jadu@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
df
```

```python
# sort by the last column 

df.sort_values(by='last',ascending=False)
```

```python
# sorting by multiple columns

df.sort_values(by=['last','first'], ascending=False)
```

```python
# sorting multiple columns on different order 

df.sort_values(by=['last','first'], ascending=[False,True])
```

```python
# sort on index 

df.sort_index()
```

```python
# sort data on series data too

df['last'].sort_values()
```
